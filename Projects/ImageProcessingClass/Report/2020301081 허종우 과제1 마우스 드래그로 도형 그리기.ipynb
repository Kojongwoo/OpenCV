```python
import numpy as np
import cv2

# 마우스 이벤트 함수
def onMouse(event, x, y, flags, param):
    global title, pt, img, copy_img

    # 왼쪽 버튼을 눌렀을 때
    if event == cv2.EVENT_LBUTTONDOWN:
        if pt[0] < 0:
            pt = (x, y)  # 시작 좌표 기록
        copy_img = img.copy()  # 임시 이미지 저장

    # 왼쪽 버튼을 누르고 이동할 때 (실시간 사각형 그리기)
    elif event == cv2.EVENT_MOUSEMOVE and flags == cv2.EVENT_FLAG_LBUTTON:
        if pt[0] >= 0:                    # 시작 좌표가 기록된 경우에만
            img = copy_img.copy()         # 이전 상태로 복원
            cv2.rectangle(img, pt, (x, y), (255, 0, 0), 2)  # 파란색 사각형 그리기
            cv2.imshow(title, img)

    # 왼쪽 버튼을 뗐을 때 (최종 사각형 그리기)
    elif event == cv2.EVENT_LBUTTONUP:
        cv2.rectangle(img, pt, (x, y), (255, 0, 0), 2)  # 최종 사각형 그리기
        cv2.imshow(title, img)
        pt = (-1, -1)               # 시작 좌표 초기화

    # 오른쪽 버튼을 눌렀을 때
    elif event == cv2.EVENT_RBUTTONDOWN:
        if pt[0] < 0:
            pt = (x, y)             # 시작 좌표 기록
        copy_img = img.copy()       # 임시 이미지 저장

    # 오른쪽 버튼을 누르고 이동할 때 (실시간 원 그리기)
    elif event == cv2.EVENT_MOUSEMOVE and flags == cv2.EVENT_FLAG_RBUTTON:
        if pt[0] >= 0:                           # 시작 좌표가 기록된 경우에만
            img = copy_img.copy()                        # 이전 상태로 복원
            dx, dy = x - pt[0], y - pt[1]                # 원의 반지름 계산
            radius = int(np.sqrt(dx**2 + dy**2))
            cv2.circle(img, pt, radius, (0, 0, 255), 2)  # 빨간색 원 그리기
            cv2.imshow(title, img)

    # 오른쪽 버튼을 뗐을 때 (최종 원 그리기)
    elif event == cv2.EVENT_RBUTTONUP:
        dx, dy = x - pt[0], y - pt[1]                # 두 좌표 간 간격
        radius = int(np.sqrt(dx**2 + dy**2))
        cv2.circle(img, pt, radius, (0, 0, 255), 2)  # 최종 원 그리기
        cv2.imshow(title, img)
        pt = (-1, -1)                                # 시작 좌표 초기화

# 흰색 배경 이지지
img = np.full((300, 500, 3), (255, 255, 255), np.uint8)

pt = (-1, -1)                         # 시작 좌표 초기화
title = "Draw Event"
cv2.imshow(title, img)
cv2.setMouseCallback(title, onMouse)  # 마우스 콜백 함수 등록
cv2.waitKey(0)
cv2.destroyAllWindows()
```
